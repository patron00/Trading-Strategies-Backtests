```
import yfinance
import pandas as pd
import numpy as np
import math
from scipy import stats
import matplotlib.pyplot as plt
from pytickersymbols import PyTickerSymbols
from datetime import datetime, timedelta
```

```
stock_data = PyTickerSymbols()
syp = pd.DataFrame(stock_data.get_stocks_by_index('S&P 500'))
```

```
stocks_tech = []
for i in range(len(syp)):
  for j in range(len(syp['industries'][i])):
    if syp['industries'][i][j] == 'Technology' or syp['industries'][i][j] == 'NASDAQ 100':
      stocks_tech.append(syp.symbol[i])
```

```
len(stocks_tech)
```

```
nasdaq_100=pd.DataFrame(stock_data.get_stocks_by_index('NASDAQ 100'))
nasdaq_100 = nasdaq_100[nasdaq_100.symbol != 'CTRP'].reset_index()

years = 10 
percentil = 99
mv_avg = 50

today = datetime.today().strftime('%Y-%m-%d')
start=(datetime.today() - timedelta(days=365*years)).strftime('%Y-%m-%d')
```

```
d = pd.DataFrame([])
for j in range(len(stocks_tech)):

  #ticker = yfinance.Ticker(nasdaq_100.symbol[j])
  d.insert(j,stocks_tech[j], yfinance.download(stocks_tech[j],period='252d')['Close'])#, interval='5m'
```

```
d
```

```
d_mv_avg = d.rolling(mv_avg).mean().dropna()
```

```
d_mv_avg_prueba = d_mv_avg[['CSCO', 'AAPL']]
```

```
d_mv_avg_prueba['dif'] = d_mv_avg_prueba['CSCO'] - d_mv_avg_prueba['AAPL']
```

```
titulos = d.columns.values[0:].tolist()
```

```
#%%time

max_perc = 0
titulo_g = 0
titulo_f = 0

matriz_rel=np.zeros(shape=(len(titulos),len(titulos)))
base_temp=d_mv_avg
vector_rel=np.zeros(shape=(len(base_temp)))
    
for h in range(0,len(titulos)):
        
    for j in range(0,len(titulos)):
            
        for i in range(0,len(base_temp)):
            if h==j:
                vector_rel[i]=(base_temp.iloc[i,h])#*10000
            else: 
                vector_rel[i]=(base_temp.iloc[i,h])-(base_temp.iloc[i,j])#*10000
        matriz_rel[h,j]=stats.percentileofscore(vector_rel,vector_rel[len(base_temp)-1])
    
  
for f in range(0,len(titulos)):
    for g in range(0,len(titulos)):
        if matriz_rel[f,g]>=percentil and titulos[g] != titulos[f]:
          if matriz_rel[f,g] >= max_perc:
            max_perc = matriz_rel[f,g]
            titulo_g = titulos[g]
            titulo_f = titulos[f]
          print(matriz_rel[f,g], "tomar posición larga en",titulos[f], " y corta en ", titulos[g])
    
dfbase=pd.DataFrame(matriz_rel)
dfbase.columns=titulos
dfbase.rename(index=dict(enumerate(titulos)),inplace=True)
    
print("Con un percentil máximo de", max_perc, "se toma posición larga en",titulo_f, " y corta en ", titulo_g)
```
