```
import yfinance
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
from pytickersymbols import PyTickerSymbols
from google.colab import files
from datetime import datetime, timedelta

limit_inf = [.02]
limit_sup = [.04]

limit_inf_neg = [-.04]
limit_sup_neg = [-.02]

risk = [3]
R = 25
period = 3

precio_min = [20]
precio_max = [35]
```

**CASO PARA DESPUES DEL OPEN!! (DESPUES DE LAS 8:30)**

```
tickers_in_range = []
for i in range(len(tickers_list)):
  df_dia = pd.DataFrame.copy(yfinance.download(tickers_list[i], period='2d')[['Open', 'Close']])
  df_min = pd.DataFrame.copy(yfinance.download(tickers_list[i], period='1d', interval='30m')['Open']) #, prepost=True
  
  #open = df_min.loc[today + ' 09:30:00-05:00']
  if len(df_dia) > 1 and len(df_min) > 0:
    open = df_min[0]
    porcentaje = (open - df_dia['Close'][-2]) / df_dia['Close'][-2]
    
    df_dia['True Open'] = open
    df_dia['%'] = porcentaje

    #print(tickers_list[i])
    #print(df_dia)

    if (df_dia['%'][-1] >= limit_inf and df_dia['%'][-1] < limit_sup and open >= precio_min and open < precio_max):
      tickers_in_range.append(tickers_list[i])
      print(tickers_list[i])
      print(df_dia)
    elif (df_dia['%'][-1] >= limit_inf_neg and df_dia['%'][-1] < limit_sup_neg and open >= precio_min and open < precio_max):
      tickers_in_range.append(tickers_list[i])
      print(tickers_list[i])
      print(df_dia)
```

```
for i in range(len(tickers_in_range)):
  df_dia = pd.DataFrame.copy(yfinance.download(tickers_in_range[i], period='2d')[['Open', 'Close']])
  df_min = pd.DataFrame.copy(yfinance.download(tickers_in_range[i], period='1d', interval='15m')['Open'])

  open = df_min[0]
  porcentaje = (open - df_dia['Close'][-2]) / df_dia['Close'][-2]
  
  df_dia['True Open'] = open
  df_dia['%'] = porcentaje

  if df_dia['%'][-1] > 0:
    stop = (df_dia['True Open'][-1] - df_dia['Close'][-2])/risk
    shares = R//stop
    print('Short', shares, 'de', tickers_in_range[i], 'a', df_dia['True Open'][-1], 'con STOP =', df_dia['True Open'][-1]+stop, 'y TARGET =', df_dia['Close'][-2])
  elif df_dia['%'][-1] < 0:
    stop = (df_dia['Close'][-2] - df_dia['True Open'][-1])/risk
    shares = R//stop
    print('Comprar', shares, 'de', tickers_in_range[i], 'a', df_dia['True Open'][-1], 'con STOP =', df_dia['True Open'][-1]-stop, 'y TARGET =', df_dia['Close'][-2])
```
